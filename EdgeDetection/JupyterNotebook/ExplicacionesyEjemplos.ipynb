La función a utilizar mas adelante `evalfis` para evaluar sistemas de inferencia difusa solo admite datos de precisión simple y de precisión doble. Por lo tanto, necesitamos convertir nuestra imagen utilizando la función `im2double`.

El algoritmo de deteccion de bordes propuesto se basa en **gradientes** para localizar los cambios bruscos de intensidad en la imagen, esta operacion se realiza tanto en el eje $x$ como en el eje $y$

Se utilizan filtros de gradientes simples, :

<center> $ G_{x} = 
\begin{bmatrix}
-1 & 1\\
\end{bmatrix}
$ 

</center>
---
<center> $ G_{y} = 
\begin{bmatrix}
-1 \\
1 \\
\end{bmatrix}
$ </center>

Para realizar el calculo, se utiliza la funcion `conv2`, una funcion que recibe dos matrices $A$ y $B$ ambos de dimensiones $m x n$ y $p x q$, respectivamente, que retorna una matriz $C$ de tamano $(m+p-1)$ x $(n+q-1)$. El calculo de cada celda de la matriz $C$ se realiza con la siguiente formula:

$$
C(j,k) = \sum_{p}\sum_{q} A(p,q)B(j−p+1,k−q+1)\\ 
$$
