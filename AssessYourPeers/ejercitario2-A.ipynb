**Ejercicio 8.** Demuestra las propiedades de abajo.

1. Para cualquier número difuso $u$ y cualquier $a,b\in \mathbb{R}$, con $a\cdot b\geq 0$, se cumple $(a+b)\cdot u=(a\cdot u)+(b\cdot u)$.
2. Para cualquier $\lambda\in\mathbb{R}$ y cualesquiera números difusos $u,v$, se cumple $\lambda\cdot (u+v)=\lambda\cdot u+\lambda\cdot v$.
3. Para cualquier $\lambda,\mu\in\mathbb{R}$ y cualquier número difuso $u$, se cumple $(\lambda\cdot \mu)\cdot u=\lambda\cdot (\mu\cdot u)$.

#### 1.
Definamos un número difuso $u$ con la siguiente notación: $u=(x_1,x_2,...,x_n)$

**Caso 1:** $a$ y $b$ son positivos:

>$(a+b)\cdot u=\left(x_1(a+b),x_2(a+b),...,x_n(a+b)\right)$

>$a\cdot u+b\cdot u=\left(x_1(a),x_2(a),...,x_n(a)\right)+\left(x_1(b),x_2(b),...,x_n(b)\right)$
>
>$a\cdot u+b\cdot u=\left(x_1(a)+x_1(b),x_2(a)+x_2(b),...,x_n(a)+x_n(b)\right)$
>
>$a\cdot u+b\cdot u=\left(x_1(a+b),x_2(a+b),...,x_n(a+b)\right)$

**Caso 2:** $a$ y $b$ son negativos:
>$(a+b)\cdot u=\left(x_n(a+b),x_{n-1}(a+b),...,x_1(a+b)\right)$

>$a\cdot u+b\cdot u=\left(x_n(a),x_{n-1}(a),...,x_1(a)\right)$
>
>$a\cdot u+b\cdot u=\left(x_n(a)+x_n(b),x_{n-1}(a)+x_{n-1}(b),...,x_1(a)+x_1(b)\right)$
>
>$a\cdot u+b\cdot u=\left(x_n(a+b),x_{n-1}(a+b),...,x_1(a+b)\right)$

Como se cumple para ambos casos, queda demostrado que $(a+b)\cdot u=(a\cdot u)+(b\cdot u)$

Definamos un número difuso $v$ con la siguiente notación: $v=(y_1,y_2,...,y_n)$
