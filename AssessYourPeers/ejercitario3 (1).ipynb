**Ejercicio 6.** Implementa una subrutina que toma como entrada una matriz que represente una relación difusa y retorna 1 si la matriz es una relación de similaridad y retorna 0 si no es una relación de similaridad. Escribe al menos dos ejemplos diferentes que muestren como funciona tu algoritmo. Puedes elegir tu lenguaje de programación favorito para este ejercicio.

Suponga que A es un conjunto difuso, cuyo universo es el conjunto X. Entonces, S es una relación de similaridad en A si:
1. $R(a,b) = R(b,a)$
2. $R(a,b) \leq min[R(a,a), R(b,b)],$
3. $R(a,c) ≥ ⋁_{ b∈A} $ $[R(a,b) ∗ R(b,c)]$

para toda a y b en X


**Ejemplo 1**
\begin{equation}
    R =
    \begin{bmatrix}
    a & b & c &  \\
    1.0 & 0.2 & 0.6 & a \\
    0.2 & 1.0 & 0.2 & b \\
    0.6 & 0.2 & 1.0 & c
    \end{bmatrix}
\end{equation}

**Ejemplo 2**

    \begin{bmatrix}

    0.7 & 1.0 & 0.7 & b \\
    0.6 & 0.7 & 1.0 & c
    \end{bmatrix}
\end{equation}
1. Simetrica
2. Reflexiva
3. No transitiva



```python
def rel_similaridad():
    import numpy as np

    # Lectura de la dimension de la matriz
    m = int(input("Ingrese la dimensión m: "))
    n = int(input("Ingrese la dimensión n: "))

    # Declaración de la matriz
    S = np.empty( (m, n) )
    
    # Lectura de la matriz         
    print ("Ingrese los elementos de la matriz S: ")
    for i in range(m):
        for j in range(n):
            S[i, j] = float(input())

    # Imprime la matriz
    print ("Matriz S: ")
    print ( S )

    if m == n:
        # Verificar reflexividad
        reflexivo = True
        for a in range(m):
            #global reflexivo
            reflexivo = reflexivo and S[a, a] == 1.0

        # Verificar antisimetria
        antisimetrico = True
        for a in range(m):
            for b in range(m):
                if (S[a, b] > 0 and S[b, a] > 0):
                    antisimetrico = antisimetrico and a == b

        # Verificar transitividad
        transitivo = True

        for a in range(m):
            for c in range(m):
                #global transitivo
                sup = np.array([min(S[a, b], S[b, c]) for b in range(m)]).max()
                transitivo = transitivo and S[a, c] >= sup
    
        if (reflexivo == True) and (transitivo == True) and (antisimetrico == False):
            print("1")
        else:
            print("0")
    
    else:
        print("0")
        
rel_similaridad()
```

**Ejercicio 8.** Construye un ejemplo concreto de una relación difusa $R:A\times A\to[0,1]$ infinita que sea reflexiva, simétrica pero no sea transitiva.

\begin{equation}
    R =
    \begin{bmatrix}
    a & b & c &  \\
    1.0 & 0.7 & 0.6 & a \\
    0.7 & 1.0 & 0.7 & b \\
    0.6 & 0.7 & 1.0 & c
    \end{bmatrix}
\end{equation}

1. Simetria: $R(a,b) = R(b,a)$
2. Transitividad: $R(a,c) ≥ ⋁_{ b∈A} $ $[R(a,b) ∗ R(b,c)]$
3. Reflexividad: la diagonal superior es $1.0$ en todas sus relaciones

```python
import numpy as np

S = np.array([ 
    [1.0, 0.7, 0.6],
    [0.7, 1.0, 0.7],
    [0.6, 0.7, 1.0]
])

# Verificar reflexividad
reflexivo = True

for a in range(3):
    global reflexivo
    reflexivo = reflexivo and S[a, a] == 1.0
    
print("Es reflexivo" if reflexivo else "No es reflexivo")

# Verificar antisimetria
antisimetrico = True

for a in range(3):
    for b in range(3):
        if (S[a, b] > 0 and S[b, a] > 0):
            antisimetrico = antisimetrico and a == b
            
print("No es simetrico" if antisimetrico else "Es simetrico")

# Verificar transitividad
transitivo = True

for a in range(3):
    for c in range(3):
        global transitivo
        sup = np.array([min(S[a, b], S[b, c]) for b in range(3)]).max()
        transitivo = transitivo and S[a, c] >= sup
        
print("Es transitivo" if transitivo else "No es transitivo")
```

```python
import numpy as np

S = np.array([ 
    [1.0, 0.2, 0.6],
    [0.0, 1.0, 0.2],
    [0.0, 0.0, 1.0]
])

# Verificar reflexividad
reflexivo = True

for a in range(3):
    global reflexivo
    reflexivo = reflexivo and S[a, a] == 1.0
    
print("Es reflexivo" if reflexivo else "No es reflexivo")

# Verificar transitividad
transitivo = True

for a in range(3):
    for c in range(3):
        global transitivo
        sup = np.array([min(S[a, b], S[b, c]) for b in range(3)]).max()
        transitivo = transitivo and S[a, c] >= sup
        
print("Es transitivo" if transitivo else "No es transitivo")

# Verificar antisimetria
antisimetrico = True

for a in range(3):
    for b in range(3):
        if (S[a, b] > 0 and S[b, a] > 0):
            antisimetrico = antisimetrico and a == b
            
print("No es simetrico" if antisimetrico else "Es simetrico")
```
