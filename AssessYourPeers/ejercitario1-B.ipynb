**Ejercicio 1.** Definimos un poset $(\mathbb{R}^n,\leq)$  donde para todo $x=(x_i)_{i=1}^n,y=(y_i)_{i=1}^n\in \mathbb{R}$ tenemos que $x\leq y$ si y solo si $x_i\leq y_i$ para todo $i=1,\dots,n$. Demuestra que $(\mathbb{R}^n,\leq)$ es de hecho un poset.

**Reflexivad:** Sea $a=(a_i)_{i=1}^n\in\mathbb{R}$ como se cumple que $a_i\leq a_i$ para todo $i=1,\dots,n$ (ya que cada $a_i$ pertenece a $\mathbb{R}$) tenemos que el conjunto $(\mathbb{R}^n,\leq)$ cumple con la propiedad Reflexiva.

**Transitividad:** Sean $a=(a_i)_{i=1}^n,b=(b_i)_{i=1}^n,c=(c_i)_{i=1}^n\in\mathbb{R}$. Se cumple que, por extensión de las propiedades de los números reales que para todo $i=1,\dots,n$, si $a_i\leq b_i$ y $b_i\leq c_i$, entonces $a_i\leq c_i$. Por lo tanto, se cumple la propiedad transitiva.

**Antisimetrica:** Sean $a=(a_i)_{i=1}^n,b=(b_i)_{i=1}^n\in\mathbb{R}$. Como todo los elementos pertenecen a $\mathbb{R}$, se cumple que, para todo $i=1,\dots,n$, si $a_i\leq b_i$ y $b_i\leq a_i$, entonces $a=b$. Así se cumple la propiedad antisimétrica.

Como las tres propiedades se cumplen, queda demostrado que $(\mathbb{R}^n,\leq)$ es un poset.

**Ejercicio 3.** Demostrar que $([0,1],\leq)$ es un retículo completo y distributivo.

La definición de retículo dice que dos elementos cualquiera poseen un ínfimo y supremo. Para que un retículo sea completo, cualquier subconjunto no vacío debe poseer un ínfimo y supremo. Es fácil ver que los elementos del conjunto de los subconjuntos $\{0\},\{1\},\{0,1\}$ cumplen con estas definiciones.

Para que el retículo $P$ sea distributivo se requiere que para todo $x,y,z\in P$ se cumple que:

$x\wedge (y\lor z)=(x\wedge y)\lor (x\wedge z)$

Tomemos a $x=\{0\}, y=\{1\}, z=\{0,1\}$. Entonces tenemos:

$x\wedge (y\lor z)=\{0\} \wedge (\{1\}\lor \{0,1\})=\{0\} \wedge \{1\}=\{0\}$  

$(x\wedge y)\lor (x\wedge z)=(\{0\}\wedge \{1\})\lor (\{0\}\wedge \{0,1\})=\{0\}\lor \{0\}=\{0\}$

$x\wedge (y\lor z)=(x\wedge y)\lor (x\wedge z)=\{0\}$

Por lo tanto, se cumple la propiedad distributiva.
