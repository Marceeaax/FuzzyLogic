**Ejercicio 1.** Definimos un poset $(\mathbb{R}^n,\leq)$  donde para todo $x=(x_i)_{i=1}^n,y=(y_i)_{i=1}^n\in \mathbb{R}$ tenemos que $x\leq y$ si y solo si $x_i\leq y_i$ para todo $i=1,\dots,n$. Demuestra que $(\mathbb{R}^n,\leq)$ es de hecho un poset.


$(\mathbb{R}^n,\leq)$ es un poset, ya que cumple con las siguientes condiciones: 
    
**1 - Reflexivo:** 
    
Dada una n-tupla  $x=(x_i)_{i=1}^n \in \mathbb{R}^n$ tenemos que 

($x_i \leq x_i ) ,  \forall i    \in \{1,2,....,n \} $ 

Ya que $(x_i \leq x_i)$  $ \Leftrightarrow$  $(x \leq x)$

Y si ($x_i \leq x_i )$ se cumple $(x \leq x)$

*Es reflexivo*


**2- Transitivo:**
    
Dado las n-tuplas $y=(y_i)_{i=1}^n, x=(x_i)_{i=1}^n ,z=(z_i)_{i=1}^n \in  \mathbb{R}^n$

($x_i \leq y_i ) ,  \forall i     \in \{1,2,....,n \}    \Leftrightarrow  (  x \leq y )$

($y_i \leq z_i ) ,  \forall i     \in \{1,2,....,n \}    \Leftrightarrow  (  y \leq z )$


(($x_i \leq y_i) \land (y_i \leq z_i)) \Rightarrow (x_i \leq z_i)$
 

Ya que $(x_i \leq z_i)$  $ \Leftrightarrow$  $(x \leq z)$

Entonces, si  $(x_i \leq z_i)$ se cumple $(x \leq z)$
 
 
 Por lo tanto, se cumple que
 $ x \leq y, y \leq z   \Rightarrow  x \leq  z$

*Es transitivo*

**3- Antisimetrico**
  
Dadas las siguientes n-tuplas $x=(x_i)_{i=1}^n, y=(y_i)_{i=1}^n \in  \mathbb{R}^n$ tenemos que:


($x_i \leq y_i ) ,  \forall i     \in \{1,2,....,n \}    \Leftrightarrow  (  x \leq y )$

($y_i \leq z_i ) ,  \forall i     \in \{1,2,....,n \}    \Leftrightarrow  (  y \leq x )$


(($x_i \leq y_i) \land (y_i \leq x_i)) \Rightarrow (x_i = y_i)$

Ya que $(x_i = y_i)$  $ \Leftrightarrow$  $(x = y)$

Y si $(x_i = y_i)$   entonces   $(x = y)$

*Es antisimetrico*


*Es transitivo, reflexivo y antisimetrico, se concluye entonces que $(\mathbb{R}^n,\leq)$ es un poset.*

**1- De Asociacion**

$A$ $\wedge$ ($B$ $\wedge$ $C$) $=$ ($A$ $\wedge$ $B$) $\wedge$ $C$

**Empiricamente**

Sea x $\in$ $X$. Luego,

**1.1-Caso 1:** $A$ $\leq$ $B$ y $B$ $\leq$ $C$

$A$ $\wedge$ ($B$ $\wedge$ $C$ $=$ min$\{A(x) , (B \wedge C )(x) \}$
              
= min$\{A(x) , min\{B(x),C(x)\}\}$ = min $\{ A(x), B(x)\}$ = $A(x)$ .......... (1)
              
($A$ $\wedge$ $B$) $\wedge$ $C$ =   min$\{(A \wedge B) (x) , C(x)\}$

=  min$\{ min\{A(x), B(x)\} , C(x) \}$ = min$\{A(x), C(x)\} = $A(x)$ .........(2)
              
Entonces, (1) = (2)
              
**1.2-Caso 2:** $B$ $\leq$ $A$ y  $A$ $\leq$ $C$ 
              
$A$ $\wedge$ ($B$ $\wedge$ $C$) = $B(x)$ ......... (3)
            
($A$ $\wedge$ $B$) $\wedge$ $C$ = $B(x)$...........(4)
              
Entonces, (3) = (4)
              
**1.3-Caso 3:** $C$ $\leq$ $A$ y $A$ $\leq$ $B$

($A$ $\wedge$ $B$) $\wedge$ $C$ = $C(x)$...........(6)
              
Entonces, (5) = (6)             
              
              
Y asi respectivamente, sin importar el orden de los conjuntos con el operador $\leq$ el resultado de $A$ $\wedge$ ($B$ $\wedge$ $C$) con ($A$ $\wedge$ $B$) $\wedge$ $C$  sera el mismo.
