$$\begin{aligned}
{}^\alpha\!(A\cdot B) &={}^\alpha\!A\cdot {}^\alpha\!B\\
                      &=[2\alpha-1,3-2\alpha]\cdot [2\alpha+1,5-2\alpha]\\
                      &=[\min((2\alpha-1)(2\alpha+1),(2\alpha-1)(5-2\alpha),(3-2\alpha)(2\alpha+1)),(3-2\alpha)(5-2\alpha)),\max((2\alpha-1)(2\alpha+1),(2\alpha-1)(5-2\alpha),(3-2\alpha)(2\alpha+1))].
\end{aligned}$$
Aquí se utilizó la regla de multiplicación de intervalos.

Podemos simplificar el intervalo del corte alfa como sigue. Primero definimos funciones simbólicas.

```python
a1=2*y-1; a2=3-2*y; b1=2*y+1; b2=5-2*y
c1(y)=min_symbolic(a1*b1,a1*b2,a2*b1,a2*b2)
c2(y)=max_symbolic(a1*b1,a1*b2,a2*b1,a2*b2)
```

```python
c1(y).full_simplify()
```
